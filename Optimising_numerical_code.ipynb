## **What is optimising numerical code and why should you care?**
Optimising numerical code refers to the practice of writing Python programs that perform numerical computations efficiently. It involves using strategies such as vectorised operations, ust-in-timecompilation, and profiling to reduce computation time, memory usage, and unnecessary overhead.The goal is to make code run faster while maintaining correctness.

## **Why this matters?**
Problem it solves:
Python loops are slow for large numerical datasets. As data grows, computationtime increases drastically if inefficient code is used.
> What you'll gain:
  * Faster computations, especially for large datasets.
  * Cleaner, more readable code by using vectorised patterns.
  * The ability to profile and optimise real-world applications.

## **Concept A: Vectorized patterns**
Definition:
A vectorised pattern replaces explicit loops with operations on entire arrays, leveraging optimized C and Fortran code under the hood in NumPy.

### **Key characteristics:**
* Broadcasting:
Automatically aligns arrays of different shapes for operations.
* Memory efficiency: Avoids Python-level loops and temporary objects.
* Speed: Uses low-level implementations for faster computation

```python
# example for vectorized patterns
import numpy as np
a = np.array([1,2,3,4])
b = np.array([10,20,30,40])

# vectorized addition
c = a + b
print(c)
```

## **Concept B: Cython/Numba Primer**
Definition: Cython and Numba are tools to compile Python code for speed.
* Cython: Converts Python code to C for faster execution.
* Numba: Uses just-in-time (JIT) compilation to optimize numerical functions at runtime.

### **Key characteristics:**
* Cython: Static typing, explicit compilation, generates .c files.
* Numba: Dynamic, minimal code change, uses @jit decorator.
* Targeted acceleration: Works well for math-heavy loops.

```python
# example for cython and numba primer
from numba import jit

@jit
def sum_squares(n):
  """Calculates the sum of
   squares upto n"""
  total = 0
  for i in range(n):
    total += i ** 2
  return total

print(sum_squares(10))
```

## **How vectorized patterns and JIT work together**
Vectorisation removes Python loops where possible, while JIT optimizes the remaining loops. Think of vectorisation as a first-level speedup and JIT as a second-level accelerator. Together, they produce near C-level performance in Python code.

```python
# Example 1: vectorized sum of squares

# step 1: create an np array
arr = np.arange(1,1000001)

# step 2: compute sum of squares vectorized =
result = np.sum(arr ** 2)

# Result
print(result)
```

```python
# Example 2: Using numba to speed loops
# aaply @jit from numba to accelarate the function

@jit
def sum_squares_loop(n):
  total = 0
  for i in range(1, n+1):
    total += i**2
  return total
print(sum_squares_loop(1000000))
```

#### What's happening in the above example?
Numba compiles the Python loop to machine code at runtime, so the loop executes at near C speed.

#### Key takaway:
JIT is a quick way to speed up numeric loops that cannot be vectorised.

```python
# Example 3: Broadcasting for efficient array operations
A = np.arange(1000).reshape(1000,1)
B = np.arange(500).reshape(1,500)

# Broadcasting addition
C = A + B

print(C.shape)
print(C)
```

```python
# Example 4: Combining vectorization with numba

@jit
def complex_transform(arr):
  res = np.zeros_like(arr)
  for i in range(arr.size):
    res[i] = np.sin(arr[i] ** 2 + np.log1p(arr[i]))

  return res

arr = np.arange(1, 1000000)
result = complex_transform(arr)
print(result)
```

#### What's happening in the above example?
Numba accelerates the loop, while NumPy functions inside the loop(np.sin, np.log1p) remain efficient.

#### **Key takeaway:**
Combining vectorisation with JIT provides flexibility and speed for complex computations.

## **Concept C: Profiling**
Profiling means measuring how long different
paths of your code take to run and how much memory or
cpu they use.
* %timeit : Performs quick performance checks in the code.
* cprofile: shows how much time each function in your program takes.

```python
# using %timeit
arr = np.arange(1000000)


# for loop in python:
def squares(a):
  result = []
  for x in a:
    result.append(x ** 2)
  return result

# NumPy:
def square_with_numpy(a):
  np.square(a)
```

```python
%timeit squares(arr)
%timeit square_with_numpy(arr)
```

```python
# using cprofile
import cProfile # cProfile check

def compute():
  x = np.random.rand(1000, 1000)
  y = np.dot(x, x.T) # dot product of two arrays x and x.T (Transpose of x)
  z = np.linalg.inv(y[:100,:100]) # linear algebra --> inverse of y array --> will create a slice of y of 100x100 matrix
  return z.sum()
```

```python
cProfile.run('compute()')
```

#### **using time module**

```python
# using Time module:
import time
arr = np.arange(1000000)

start = time.time()
result = np.square(arr)
end = time.time()

print(f" Execution time: {end - start:.5} seconds")
```

```python
from line_profiler import LineProfiler
import numpy as np

def matrix_ops():
  a = np.random.rand(1000, 1000)
  b = np.random.rand(1000, 1000)
  c = np.dot(a,b)
  d = np.linalg.svd(c)
  return d

lp = LineProfiler()
lp.add_function(matrix_ops)
lp.run('matrix_ops()')
lp.print_stats()
```
